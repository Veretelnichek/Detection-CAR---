```python
import cv2
import numpy as np
from pathlib import Path

def resize_to_640x640(image_path, output_path, fill_color=(114, 114, 114)):
    img = cv2.imread(str(image_path))
    if img is None:
        return False
    
    h, w = img.shape[:2]
    result = np.full((640, 640, 3), fill_color, dtype=np.uint8)
    
    scale = min(640 / w, 640 / h)
    new_w = int(w * scale)
    new_h = int(h * scale)
    
    resized = cv2.resize(img, (new_w, new_h), interpolation=cv2.INTER_LINEAR)
    
    x_offset = (640 - new_w) // 2
    y_offset = (640 - new_h) // 2
    
    result[y_offset:y_offset+new_h, x_offset:x_offset+new_w] = resized
    cv2.imwrite(str(output_path), result)
    
    return True

def batch_resize(input_dir, output_dir):
    input_path = Path(input_dir)
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)
    
    image_extensions = {'.jpg', '.jpeg', '.png', '.bmp', '.tiff', '.webp'}
    image_files = [f for f in input_path.glob("*") if f.suffix.lower() in image_extensions]
    
    for img_path in image_files:
        save_path = output_path / f"{img_path.stem}.jpg"
        if resize_to_640x640(img_path, save_path):
            print(f"{img_path.name}")
        else:
            print(f"{img_path.name}")

# Использование
batch_resize("data/upload/2", "data_resized_640")
```
